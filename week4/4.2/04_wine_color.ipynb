* Θα χρησιμοποιήσουμε (προσαρμοσμένο) ένα σύνολο δεδομένων που αφορά [κρασιά από την Πορτογαλία](https://archive.ics.uci.edu/ml/datasets/wine+quality).

* Εμείς θέλουμε, με βάση τα χαρακτηριστικά ενός κρασιού, να μπορούμε να προβλέψουμε το χρώμα του (1 = κόκκινο, 0 = λευκό).

```python
import pandas as pd
wines = pd.read_csv("wines_color.csv")
wines
```

* Θα βάλουμε τα χαρακτηριστικά και τον στόχο σε δύο χωριστούς πίνακες.

```python
X = wines.loc[:,'fixed acidity':'alcohol'].values
y = wines['color'].values
```

* Θα χρησιμοποιήσουμε μία ακόμα διαφορετική μέθοδο για ταξινόμηση, η οποία βασίζεται στο Θεώρημα του Bayes.

* Όπως έχουμε πει, το καλό με το scikit-learn είναι ότι πολύ εύκολα μπορούμε να μετακινούμαστε από μία μέθοδο σε μία άλλη.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score

clf = GaussianNB()
scores = cross_val_score(clf, X, y, cv=5)
print(scores)
print(scores.mean(), scores.std())
```

* Τώρα, ας δούμε τι θα συμβεί αν χρησιμοποιήσουμε PCA ως πρώτο βήμα πριν την ταξινόμηση.

```python
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA

clf = make_pipeline(PCA(n_components=2), GaussianNB())
scores = cross_val_score(clf, X, y, cv=5)
print(scores)
print(scores.mean(), scores.std())
```

* Έχουμε βεβαίως πει ότι πριν από PCA καλό είναι να προσαρμόζουμε τα δεδομένα μας ώστε να έχουν μέση τιμή 0 και διακύμανση 1.

* Ο μηχανισμός της διοχέτευσης μας βοηθά στο να εξασφαλίσουμε ότι θα γίνουν όλα με τη σειρά, σωστά, χωρίς διαρροή πληροφορίας από την εκπαίδευση στον έλεγχο.

* Όπως βλέπουμε, μετά την προβολή στις δύο διαστάσεις παίρνουμε πλέον καλύτερα αποτελέσματα ακόμα και από την ταξινόμηση στις πλήρεις διαστάσεις!

```python
from sklearn.preprocessing import StandardScaler

clf = make_pipeline(StandardScaler(), PCA(n_components=2), GaussianNB())
scores = cross_val_score(clf, X, y, cv=5)
print(scores)
print(scores.mean(), scores.std())
```

* Εδώ ας σταθούμε λίγο να δούμε τι πετύχαμε.

* Από ό,τι φαίνεται στις δύο διαστάσεις είναι πιο εύκολο να ταξινομηθούν τα κρασιά.

* Επίσης, στις δύο διαστάσεις είναι πιο εύκολο να δούμε τι συμβαίνει, σε σχέση με τις αρχικές 11 διαστάσεις.

* Ας το επιβεβαιώσουμε, ξεκινώντας από την αρχή.

* Χρησιμοποιούμε `StandardScaler` στα δεδομένα μας.

* Το εφαρμόζουμε σε όλα τα δεδομένα αφού δεν μας ενδιαφέρει αυτή τη στιγμή ο διαχωρισμός σε εκπαίδευση και έλεγχο.

```python
standard_scaler = StandardScaler()
X_scaled = standard_scaler.fit_transform(X)
X_scaled
```

* Ελαττώνουμε τις διαστάσεις σε δύο.

```python
pca = PCA(n_components=2)
X_transformed = pca.fit_transform(X_scaled)
X_transformed
```

* Και απεικονίζουμε τις δύο διαστάσεις και το χρώμα του κάθε κρασιού.

* Μπορούμε να καταλάβουμε γιατί δούλεψε καλά η ταξινόμηση: τα κρασιά μπορούν να ενταχθούν, στις δύο διαστάσεις, σε δύο διακριτές συστάδες.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(x=X_transformed[:,0], y=X_transformed[:,1], c=wines['color'], edgecolor='k', cmap='Reds')
plt.grid()
```
